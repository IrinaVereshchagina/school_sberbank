```
pd.read_sql_query("""
SELECT fio, job, extract(month from date_n) AS "месяц", salary_ver.fine

FROM worker_ver

INNER JOIN executor_ver on executor_ver.kod_work=worker_ver.id

INNER JOIN salary_ver on salary_ver.kod_work=worker_ver.id

WHERE (date_n-date_b)>7 AND salary_ver.fine<100



  """, conn)
#если разница между датой начала работы и датой окончания то данный сотрудник в следующем периоде может получить не 100% премии а 50, 
```

```
pd.read_sql_query("""
SELECT so_ver.name, service_ver.name, so_ver.fio, so_ver.tel, (dogovor_ver.sum-pay_ver.sum) AS "Задолженность"

FROM dogovor_ver

INNER JOIN so_ver on so_ver.id=dogovor_ver.kod_set

INNER JOIN pay_ver on pay_ver.kod_dog=dogovor_ver.kod_dog AND pay_ver.kod_serv=dogovor_ver.kod_serv

INNER JOIN service_ver on service_ver.id=dogovor_ver.kod_serv

WHERE (dogovor_ver.sum-pay_ver.sum) >0

ORDER BY dogovor_ver.kod_set


  """, conn)
#вывод суммы задолженности с указанием организации, данных директора и телефона
```

```
pd.read_sql_query("""
SELECT name, price
FROM service_ver
INNER JOIN dogovor_ver ON dogovor_ver.kod_serv=service_ver.id

WHERE dogovor_ver.kol>1


  """, conn)
#выбираем сервисы у которых колличество периодов использования больше одного
```
